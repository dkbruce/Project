```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
mw_data = "Resources/midwest.csv"
s_data = "Resources/south.csv"
ne_data = "Resources/northeast.csv"
w_data = "Resources/west.csv"
```

```python
# Read data and store into a data frame for sex
midwest_suicide_df = pd.read_csv(mw_data)
south_suicide_df = pd.read_csv(s_data)
northeast_suicide_df = pd.read_csv(ne_data)
west_suicide_df = pd.read_csv(w_data)
```

```python
full_data_3 = midwest_suicide_df.append(west_suicide_df)
full_data_2 = full_data_3.append(northeast_suicide_df)
full_data = full_data_2.append(south_suicide_df)
full_data
```

```python
# Set region type
# region_type = full_data.set_index("Region")

south = full_data.loc[full_data["Region"]== "South"]
west = full_data.loc[full_data["Region"]== "West"]
northeast = full_data.loc[full_data["Region"]== "Northeast"]
midwest = full_data.loc[full_data["Region"]== "Midwest"]

plt.scatter(south["Year"], south["Crude Rate"], s = south["Population"]/100000, marker="o", label = "South", c = "goldenrod", alpha = 0.85, edgecolor = "black", linewidth = .25)
plt.scatter(west["Year"], west["Crude Rate"], s = west["Population"]/100000, marker="o", label = "West", c = "indianred", alpha = 0.85, edgecolor = "black", linewidth = .25)
plt.scatter(northeast["Year"], northeast["Crude Rate"], s = northeast["Population"]/100000, marker="o", label = "Northeast", c = "cornflowerblue", alpha = 0.85, edgecolor = "black", linewidth = .25)
plt.scatter(midwest["Year"], midwest["Crude Rate"], s = midwest["Population"]/100000, marker="o", label = "Midwest", c = "darkseagreen", alpha = 0.85, edgecolor = "black", linewidth = .25)


plt.ylim = (0, west["Crude Rate"]+10)
plt.title("Regional Rates of Suicide in the US (2013 - 2017)", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Suicide Rate (%)")

plt.text(10, 2, "Marker size reflects population size")
plt.legend(bbox_to_anchor=(0.001, 1, 1, -1.2), loc=7, title="Note: Marker size reflects population size", title_fontsize = 10, fontsize = 8, ncol=4, markerscale=.5, borderpad = 1.5)
# plt.legend(bbox_to_anchor=(0., 1.02, 1, .102), loc=2,
#            ncol=4, mode="expand", borderaxespad=0.)
plt.tight_layout
plt.grid()
plt.show()
```
