```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
mw_data = "Resources/midwest.csv"
s_data = "Resources/south.csv"
ne_data = "Resources/northeast.csv"
w_data = "Resources/west.csv"
```

```python
# Read data and store into a data frame for sex
midwest_suicide_df = pd.read_csv(mw_data)
south_suicide_df = pd.read_csv(s_data)
northeast_suicide_df = pd.read_csv(ne_data)
west_suicide_df = pd.read_csv(w_data)
west_suicide_df.head()
```

```python
mw_rate = midwest_suicide_df["Crude Rate"]
w_rate = west_suicide_df["Crude Rate"]
s_rate = south_suicide_df["Crude Rate"]
ne_rate = northeast_suicide_df["Crude Rate"]

x = np.arange(5)
years = "2013", "2014", "2015", "2016", "2017"
width = 0.8


plt.bar(x + .3, w_rate, color="green", width = 0.2, label = "West")
plt.bar(x + .1, s_rate, color="red", width = 0.2, label = "South")
plt.bar(x - .1, mw_rate, color="blue", width = 0.2, label = "Midwest")
plt.bar(x - .3, ne_rate, color="yellow", width = 0.2, label = "Northeast")

tick_locations = [value for value in years]
plt.xticks(x, years)
plt.xlim = (-10, len(years)+.25)
plt.ylim = (0, max(w_rate)+100)

plt.title("Regiona Suicide Rates (2013 - 2017)", fontsize = 16)
plt.xlabel("Year")
plt.ylabel("Percent of Suicide Deaths (Regional)")
plt.legend(loc=(1.05, 0.4))
plt.tight_layout()

plt.show()
```
